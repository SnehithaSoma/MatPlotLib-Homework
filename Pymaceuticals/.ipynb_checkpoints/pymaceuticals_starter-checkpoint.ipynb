```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import scipy.stats as st
from sklearn import datasets

# Study data files
mouse_metadata_path = "D:\Homework\MatPlotLib-Homework\Pymaceuticals/Mouse_metadata.csv"
study_results_path = "D:\Homework\MatPlotLib-Homework\Pymaceuticals/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_metadata,study_results, how= "outer" , on= "Mouse ID" )

# Display the data table for preview

combined_data_df
                     
```

## Summary Statistics

```python
#Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

#This method is the most straighforward, creating multiple series and putting them all together at the end.

mean = combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median = combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variance= combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
standard_dev = combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sem = combined_data_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()


summary_df = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": standard_dev, "SEM": sem})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

single_groupby = clean_df.groupby('Drug Regimen')
summary_single = single_groupby.agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]
summary_single
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin


# Start by getting the last (greatest) timepoint for each mouse
cap_df = combined_data_df.loc[combined_data_df["Drug Regimen"] == "Capomulin",:]
cap_greatest= cap_df.groupby("Mouse ID").max()["Timepoint"]
cap_total_df = pd.DataFrame(cap_greatest)

ram_df = combined_data_df.loc[combined_data_df["Drug Regimen"] == "Ramicane", :]
ram_greatest= ram_df.groupby("Mouse ID").max()["Timepoint"]
ram_total = pd.DataFrame(ram_greatest)

inf_df = combined_data_df.loc[combined_data_df["Drug Regimen"] == "Infubinol", :]
inf_greatest= inf_df.groupby("Mouse ID").max()["Timepoint"]
inf_total = pd.DataFrame(inf_greatest)

ceft_df = combined_data_df.loc[combined_data_df["Drug Regimen"] == "Ceftamin", :]
ceft_greatest= ceft_df.groupby("Mouse ID").max()["Timepoint"]
ceft_total = pd.DataFrame(ceft_greatest)

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
cap_merge = pd.merge(cap_total_df,combined_data_df, on=("Mouse ID","Timepoint"),how="left")
ram_merge = pd.merge(ram_total,combined_data_df, on=("Mouse ID","Timepoint"),how="left")
inf_merge = pd.merge(inf_total,combined_data_df, on=("Mouse ID","Timepoint"),how="left")
ceft_merge = pd.merge(ceft_total,combined_data_df, on=("Mouse ID","Timepoint"),how="left")


inf_tumor= inf_merge["Tumor Volume (mm3)"]
ceft_tumor= ceft_merge["Tumor Volume (mm3)"]
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

corr_cap=round(st.pearsonr(cap_avg['Weight (g)'],cap_avg['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {corr_cap}")

x_values = cap_avg['Weight (g)']
y_values = cap_avg['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f"slope:{slope}")
print(f"intercept:{intercept}")
print(f"rvalue (Correlation coefficient):{rvalue}")
print(f"pandas (Correlation coefficient):{corr_cap}")
print(f"stderr:{stderr}")

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(line_eq)

```

```python
fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.scatter(x_values,y_values,s=175, color="blue")
plt.plot(x_values,regress_values,"r-")
plt.title('Regression Plot of Mouse Weight Versus Average Tumor Volume',fontsize =20)
plt.xlabel('Weight(g)',fontsize =14)
plt.ylabel('Average Tumore Volume (mm3)',fontsize =14)
ax1.annotate(line_eq, xy=(20, 40), xycoords='data',xytext=(0.8, 0.95), textcoords='axes fraction',horizontalalignment='right', verticalalignment='top',fontsize=30,color="red")

print(f"The r-squared is: {rvalue**2}")

plt.show()
```
